```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

plt.style.use("fivethirtyeight")
```

```python
def f(n):
    a=list(str(n))
    a.sort()
    return int("".join(a))

def test_f():
    a=defaultdict(int)
    for i in range(1,int(1e4)):
        a[f(i)]+=1
    print(a)
    # plot
    x = np.array(list(a.keys()))
    y = np.array(list(a.values()))
    plt.plot(x, y, 'o')
    plt.show()

test_f()
```
